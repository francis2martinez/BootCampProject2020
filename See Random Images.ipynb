```python
#Import packages
import pandas as pd
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
Catalog_train=pd.read_pickle('Catalog_train.pkl')
```

```python
Catalog_train.sample(4)
```

```python
#Finally Let's see some of the pictures.
#taken from https://www.bogotobogo.com/python/OpenCV_Python/python_opencv3_matplotlib_rgb_brg_image_load_display_save.php
def plot_image(image,ax):
    b,g,r = cv2.split(image)       # get b,g,r
    rgb_img = cv2.merge([r,g,b])     # switch it to rgb

    ax.imshow(rgb_img)
    ax.set_xticks([]); ax.set_yticks([])   # to hide tick values on X and Y axis
```

```python
def show_random_pictures(n,data=Catalog_train):
    data=Catalog_train
    images_vangogh=list(data.loc[data.Vangogh==1,['Image']].sample(n).Image)
    images_others=list(data.loc[data.Vangogh==0,['Image']].sample(n).Image)
    fig,axes=plt.subplots(n,2,figsize=(15,8*n))
    axes[0,0].set_title('Van Gogh'); axes[0,1].set_title('Others')
    for i in range(n):
        plot_image(images_vangogh[i],axes[i,0])
        plot_image(images_others[i],axes[i,1])
```

```python
show_random_pictures(5)
```
